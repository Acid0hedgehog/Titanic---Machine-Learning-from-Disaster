```python
import pandas as pd
import numpy as np
```

```python
gender = pd.read_csv('gender_submission.csv')
gender
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train_copy = train.copy()
```

```python
train_copy
```

```python
train_copy = train_copy.drop('Name', axis=1)
train_copy
```

```python
train_copy = train_copy.drop('Cabin', axis=1)
train_copy
```

```python
train_copy = train_copy.drop('Ticket', axis=1)
train_copy
```

```python
train_copy['Sex'] = train_copy['Sex'].replace('male', '1')
train_copy['Sex'] = train_copy['Sex'].replace('female', '0')
train_copy
```

```python
train_copy['Embarked'] = train_copy['Embarked'].replace('S', '1')
train_copy['Embarked'] = train_copy['Embarked'].replace('C', '2')
train_copy['Embarked'] = train_copy['Embarked'].replace('Q', '3')
train_copy
```

```python
X = train_copy.drop('PassengerId', axis=1)
X = X.drop('Survived', axis=1)
X
```

```python
avg = X['Age'].mean()
avg = round(avg,2)
avg
```

```python
X['Age'].fillna(avg, inplace=True)
X
```

```python
X['Embarked'].fillna(0, inplace=True)
X
```

```python
Y = train_copy['Survived']
Y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
```

```python
model.fit(X, Y)
```

```python
test_copy = test.copy()
test_copy = test_copy.drop('Name', axis=1)
test_copy = test_copy.drop('Cabin', axis=1)
test_copy = test_copy.drop('Ticket', axis=1)
test_copy['Sex'] = test_copy['Sex'].replace('male', '1')
test_copy['Sex'] = test_copy['Sex'].replace('female', '0')
test_copy['Embarked'] = test_copy['Embarked'].replace('S', '1')
test_copy['Embarked'] = test_copy['Embarked'].replace('C', '2')
test_copy['Embarked'] = test_copy['Embarked'].replace('Q', '3')
test_copy
```

```python
gender
```

```python
X_test = test_copy.drop('PassengerId', axis=1)
X_test
```

```python
avg = X_test['Age'].mean()
avg = round(avg,2)
X_test['Age'].fillna(avg, inplace=True)
X_test['Embarked'].fillna(0, inplace=True)
X_test
```

```python
rows_with_nan = X_test[X_test.isna().any(axis=1)]
rows_with_nan
```

```python
avg = X_test['Fare'].mean()
avg = round(avg,2)
X_test['Fare'].fillna(avg, inplace=True)
X_test
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
gender['y_pred'] = y_pred
gender
```

```python
gender.rename(columns={'y_pred': 'Survived'}, inplace=True)
gender
```
